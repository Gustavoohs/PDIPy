```python
import numpy as np
import tifffile as tiff
from spectral import imshow
import matplotlib

matplotlib.rcParams['figure.figsize'] = (15,10)
```

```python
L5_2011 = tiff.imread('LT05_L1TP_232066_20110806_T1_W.tif')
L8_2020 = tiff.imread('LC08_L2SP_232067_20200729_T1_W.tif')
```

```python
L5_NDVI = ((L5_2011[:,:,3]-L5_2011[:,:,2])/(L5_2011[:,:,3]+L5_2011[:,:,2]))
L8_NDVI = ((L8_2020[:,:,4]-L8_2020[:,:,3])/(L8_2020[:,:,4]+L8_2020[:,:,3]))
```

```python
diff = L8_NDVI - L5_NDVI
```

```python
imshow(diff, stretch=(0.02))
```

```python
import matplotlib.pyplot as plt
vdiff = diff.flatten()
plt.hist(vdiff, bins=200)
plt.text(-0.3, 26000, r'$\mu=-0.23035367$', fontsize=15)
plt.annotate('Limiar inferior = -0.435', xy=(-0.435, 1900),  xycoords='data',
            xytext=(-0.635, 10000), textcoords='data', size=15,
            arrowprops=dict(facecolor='red', shrink=0.05),
            horizontalalignment='left', verticalalignment='bottom',
            )
plt.annotate('Limiar superior = -0.026', xy=(-0.026, 1900),  xycoords='data',
            xytext=(-0.026, 10000), textcoords='data', size=15,
            arrowprops=dict(facecolor='green', shrink=0.05),
            horizontalalignment='left', verticalalignment='bottom',
            )
plt.grid(True)
plt.show()
```
