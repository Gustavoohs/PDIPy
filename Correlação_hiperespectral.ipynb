```python
import spectral.io.envi as envi 
import numpy as np  
import seaborn as sns
import pandas as pd
```

```python
#Leitura da imagem
img = envi.open('sja3.hdr', 'sja3.rfl')
img_array = np.array(img.load())
```

```python
#Transformação da imagem em tabela
Matriz = np.zeros((img_array[:,:,0].size,224))

for i in range(224):
    vetor = img_array[:,:,i].flatten()
    Matriz[:,i] = vetor 

df = pd.DataFrame(Matriz)
corr = df.corr()
```

```python
mask = np.triu(np.ones_like(corr, dtype=bool))#Máscara

sns.heatmap(corr, mask=mask)#Correlação
```
