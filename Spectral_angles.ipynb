```python
# Bibliotecas
import numpy as np
import matplotlib.pyplot as plt
import tifffile as tif
from spectral import *
```

```python
# Leitura de arquivos
img = tif.imread('L71221071_07120010720_DN.tif')
gt = tif.imread('roi.tif')
```

```python
# Criação das classes de treinamento
classes = create_training_classes(img,gt)
```

```python
# Computando média das classes
means = np.zeros((len(classes), img.shape[2]),float)

for (i, c) in enumerate(classes):
  means[i] = c.stats.mean
```

```python
# Cálculo dos ângulos
angles = spectral_angles(img, means)

angles_ord = np.argmin(angles, 2)
```

```python
# Visualização do resultado
plt.figure(figsize=(12,8))
plt.imshow(angles_ord, cmap='Spectral_r')
```
