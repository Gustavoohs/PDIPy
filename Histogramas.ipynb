```python
import matplotlib.pyplot as plt
import numpy as np
import earthpy.spatial as es
import earthpy.plot as ep
plt.style.use('dark_background')
```

```python
banda2 = "B2.tif"
banda3 = "B3.tif"
banda4 = "B4.tif"
banda5 = "B5.tif"

lista_de_bandas = [banda2,banda3,banda4,banda5]
stack, meta_data = es.stack(lista_de_bandas)
```

```python
titles = ["Green","Red", "NIR", "SWIR 1"]

colors = ["Green","Red","Maroon","Purple"]

ep.plot_bands(stack, title=titles)
plt.show()
```

```python
ep.hist(stack, bins = 100, colors=colors, title=titles)
plt.show()
```
