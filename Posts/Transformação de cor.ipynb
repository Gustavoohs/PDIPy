```python
import numpy as np
import skimage.color as color
import matplotlib.image as mpimg
import matplotlib.pyplot as plt
from osgeo import gdal
plt.style.use('dark_background')
```

```python
#Leitura das bandas
b2 = gdal.Open('B2.tif')
b3 = gdal.Open('B3.tif')
b4 = gdal.Open('B4.tif')
b5 = gdal.Open('B5.tif')

B2 = b2.ReadAsArray()
B3 = b3.ReadAsArray()
B4 = b4.ReadAsArray()
B5 = b5.ReadAsArray()

#Empilhamento
lista_de_bandas = [B2,B3,B4,B5]
stack = np.dstack(lista_de_bandas)
```

```python
ihs = color.rgb2hsv(stack[:,:,:3]) #Transformação 
```

```python
#Visualização
fig,axes = plt.subplots(1,3,figsize=(10,5),sharex='all', sharey='all') 
axes = axes.ravel() 

axes[0].imshow(ihs[:,:,2],cmap='gray') 
axes[0].set_title('Hue',fontsize=15) 

axes[1].imshow(ihs[:,:,1],cmap='gray') 
axes[1].set_title('Saturation',fontsize=15)  

axes[2].imshow(ihs[:,:,0],cmap='gray') 
axes[2].set_title('Value',fontsize=15)
plt.show()
```
