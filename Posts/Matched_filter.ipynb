```python
import matplotlib
matplotlib.rcParams['figure.figsize'] = (15,10)
```

```python
# Bibliotecas
from spectral import imshow, matched_filter
import tifffile as tif
```

```python
# Leitura da imagem
img = tif.imread('/content/drive/MyDrive/Live PDI Python/CBERS4/cut.tif')
```

```python
# Visualização
imshow(img, (2,3,1))
```

```python
# Pixel referência
t = img[582, 37]
```

```python
# Calculando scores
scores = matched_filter(img, t)
```

```python
# Visualizando resultado
imshow((scores > 0.3))
```

```python
imshow((scores > 0.9))
```
