```python
import matplotlib
matplotlib.rcParams['figure.figsize'] = (15,10)
```

```python
# Bibliotecas
import numpy as np
import matplotlib.pyplot as plt
import tifffile as tif
from sklearn.discriminant_analysis import QuadraticDiscriminantAnalysis
```

```python
# Lendo arquivos de imagem e áreas de treino
img = tif.imread('/content/L71221071_07120010720_DN.tif')
roi = tif.imread('/content/roi.tif')
```

```python
X = img[roi > 0] # Dados de treino
y = roi[roi > 0] # Classes conhecidas (Ground Truth)
```

```python
# Criando modelo
clf = QuadraticDiscriminantAnalysis(reg_param=1)
clf.fit(X, y)
```

```python
# Criando um vetor de duas dimensões para a classificação pixel a pixel
img2d= img.reshape(img.shape[0] * img.shape[1], img.shape[2])

# Classificação para cada pixel
pred2d = clf.predict(img2d)

# Retornando a classificação para o formato original da imagem
pred_final = pred2d.reshape(img[:, :, 0].shape)
```

```python
# Visualizando resultado
plt.imshow(pred_final, cmap='Spectral')
```
