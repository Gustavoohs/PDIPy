```python
from spectral import imshow
import tifffile as tif
import numpy as np
import matplotlib.pyplot as plt 
from scipy.signal import savgol_filter
```

```python
ndvi = tif.imread('C:/Users/Pichau/ndviMODIS.tif')
```

```python
plt.plot(ndvi[200,200,:], color = 'green', linewidth=1.5)
plt.grid()
```

```python
sav =savgol_filter(ndvi, 9, 3, axis=2)
```

```python
sav2 =savgol_filter(ndvi, 13, 3, axis=2)
```

```python
plt.plot(ndvi[200,200,:], color = 'green', linewidth=1, alpha=0.3)
plt.plot(sav[200,200,:], color = 'blue', linewidth=2, alpha=0.4)
plt.plot(sav2[200,200,:], color = 'red', linewidth=2)
plt.grid()
plt.legend(labels=['Original','Janela 9x9','Janela 13x13'])
plt.show()
```

```python
plt.plot(sav2[200,200,:], color = 'red', linewidth=2)
plt.grid()
plt.show()
```
